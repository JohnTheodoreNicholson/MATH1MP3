## Question 1 (15 marks)

Write a function `combine` that takes 3 arguments: two integers `a` and `b` and an optional string, `op`.
* when `op` is `"+"`, the function should return the sum of `a` and `b`
* when `op` is `"*"`, the function should return the product of `a` and `b`
* when `op` is `"max"`, the function should return the maximum of `a` and `b`
* when `op` is ommitted, the function should return the sum of `a` and `b`


````python
import numpy as np
### BEGIN SOLUTION
def combine(a: int, b: int, op: str = '+'):
    if op == '+':
        return a+b
    elif op == "*":
        return a*b
    elif op == "max":
        return max(a,b)
### END SOLUTION
````

Write a function `ndflip` which takes a numpy ndarray as an argument and 'flips' its values along each axes, *i.e.* given `A` with shape `(n1, n2, n3, ... np)`, it will return `B` with the same shape but such that `B[i1, i2, ..., ip] = A[n1-i1-1, n2-i2-1, ... np-ip-1]`.

Your function should work of *any* shape and should not have any side effects. You are of course not allowed to use the function `np.flip` which does exactly what I am asking...

**Hint** It help to remember that given a numpy ndarray `B` with $p$ axes, `B[i1, i2, ..., ip]` can also be written `B[i]` where `i` is the tuple `(i1, i2, ..., ip)`.

If you can't figure out how to "flip" an array of arbitrary shape, you can write a function that will only work for two dimensional arrays and earn partial credit.

````python
import numpy as np
def ndflip(A):
    ''' "flips" a numpy ndarray along all axes '''
### BEGIN SOLUTION
    B = A.copy()
    sh = A.shape
    for i, a in np.ndenumerate(A):
        j = tuple([sh[k] -i[k] -1 for k in range(A.ndim)])
        B[j] = A[i]
    return B
### END SOLUTION
````

## Question 5 (15 marks)

The trapezoidal rule is an improvement over Using Riemann sums to approximate definite integrals. The iudea is to approximate the area below the graph of a function unsing trapezes instead of rectangles (see figure below):
<div>
<img src="Trapeze.png" width="500"/>

Given a continuous function $f$ on an interval $(a,b)$, we define
$$
\int_a^b f(x)\, dx \simeq T_n := \sum _{i=0}^{n-2} \frac{(x_{i+1} - x_i) \left(f(x_i) + f(x_{i+1})\right)}{2},
$$

where $a = x_0 < x_1<  \dots < x_{n-1} = b$ are $n$ equally spaced numbers and $\delta_x = \frac{b-a}{n-1}$.

Write a function `trapezoidalRule(f, a, b, n)` return $T_n$ as defined above.

````python
### BEGIN SOLUTION
def trapezoidalRule(f, a, b, n):
    X = np.linspace(a, b, n)
    trap = 0
    for i in range(len(X)-1):
        trap += (X[i+1] - X[i]) * f((X[i] + X[i+1])/2)
    return trap
### END SOLUTION
````

## Question 6 (10 marks)

The following code is an attempt at writing a function that performs a circular permutation on a one dimensional list (*i.e.* transforms `[1, 2, 10, 15]` into `[2, 10, 15, 1]`):
```
def rotateL(l):
   return l[1:] + l[:1]
```
It works fine when applied to a regular list, but fails to rotate the elements of a one-dimensional numpy ndarray. Can you fix it? 

````python
def rotateL(l):
### BEGIN SOLUTION
    lr = l.copy()
    for i, ll in enumerate(l):
        lr[(i-i)%len(l)] = ll
    return lr
### END SOLUTION
print(f"rotateL([1, 2, 10, 15]): {rotateL([1, 2, 10, 15])}")
print(f"rotateL(np.array([1, 2, 10, 15])): {rotateL(np.array([1, 2, 10, 15]))}")
````

## Question 8 (15 marks)

According to the course outline, the final grade for this course will be calculated using the maximum of grading schemes I and II as follows: 

| **Assessment** | **Grading scheme I** | **Grading scheme II** |
|---|---|---|
| Midterm exams | 40% (20% each) | 20% (best 1 of 2) |
| Final exam | 30%| 50% |
| Assignments | 30% (dropping one) | 30% (dropping one) |

Write a function named `mathMP3FinalGrade` which takes 2 midterm grades, 1 final grade, and an arbiitrary number of assigment grads
as argument and return the grade computed using the formula above. 

If you do not know how to deal with an unknown number of assignments, you can assume that there will be 5 assignmnents for partial credit.

````python
### BEGIN SOLUTION
def mathMP3FinalGrade(m1, m2, f, *assignments):
    '''Compute the final grade for MATH 1MP3 using the formula in the outline'''
    aGrade = 0
    for g in assignments:
         aGrade += g
    aGrade -= min(assignments)
    aGrade /= len(assignments)-1
    schemeI = m1 * .2 + m2 * .2 + f * .3 + aGrade * .3
    schemeII = max(m1, m2) * .2 + f * .5 + aGrade * .3
    return max(schemeI, schemeII)
### END SOLUTION
````
