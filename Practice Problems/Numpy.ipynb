Write a function `altsign` which, given a ndarray `A` of dimension `n` returns a ndarray `B` of identical shape and such that `B[i1, i2, ..., in] = A[i1, i2, ..., in]` if `i1 + i2 + ... + in` is odd, and `B[i1, i2, ..., in] = -A[i1, i2, ..., in]` otherwise

* `altsign(np.array([1., 2., 4.]))` should return `[-1., 2., -4.]`
* `altsign(np.array([[1, 2, 3], [1, 5, 6]]))` should return `[[-1, 2, -3], [1, -5, 6]]`

Your function should work for ndarrays `A` of arbitrary dimension and shape.

```python
def alt_sign(A)
B = np.empty(A.shape)
for idx, element in np.ndenumerate(A):
    if sum(idx) %2 !=0 :
        B[idx] = element
    else:
        B[idx] = -element
return B
         
```
