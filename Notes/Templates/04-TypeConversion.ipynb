# 3 Type conversion
One unusual feature of python is that variable types are not frozen but can change through *implicit type conversion*.

```python
x = 1
print(x, type(x))
x = x/3.
print(x, type(x))
```

```python
x = 'oualala!'
print(x, type(x))
```

It is possible (but rarely necessary) to force conversion between types using the `int()`, `float()`, or `str()` operations

```python
x = str(42)
print(x, type(x))
x = float(42)
print(x, type(x))
```

```python
# this makes no sense
x = int('oulala')
```

```python
# but this does
x = '2.3'
print(x, type(x))
x = float(x)
print(x, type(x))
```

```python
x = 1
x = x + 2.3
print(x, type(x))
```

```python
x = 3
x = 'oualala ' * x 
print(x, type(x))
```
